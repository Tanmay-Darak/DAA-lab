```python
#Q1
import time
import matplotlib.pyplot as plt

def iterative_sum(N):
    total = 0
    for i in range(1, N + 1):
        total += i
    print("Sum of N numbers:",total)

def recursive_sum(N):
    if N == 0:
        return 0
    else:
        return N + recursive_sum(N - 1)

def measure_time(func, N):
    start_time = time.time()
    func(N)
    end_time = time.time()
    return end_time - start_time

# Varying N's value
N_values = list(range(1, 500, 10))  # From 1 to 1000 with step 50
iterative_times = []
recursive_times = []

# Measuring time taken for both algorithms
for N in N_values:
    iterative_times.append(measure_time(iterative_sum, N))
    recursive_times.append(measure_time(recursive_sum, N))

# Plotting the results
plt.plot(N_values, iterative_times, label='Iterative')
plt.plot(N_values, recursive_times, label='Recursive')
plt.xlabel('Value of N')
plt.ylabel('Time (seconds)')
plt.title('Time Taken for Sum of First N Natural Numbers')
plt.legend()
plt.show()
```

```python
#Q2
import random
import time
import matplotlib.pyplot as plt

# Linear search function
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

array = [random.randint(1, 1000) for _ in range(10000)]
array.sort()
key = int(input("Enter the search element"))
pos_linear = linear_search(array,key)
pos_binary = binary_search(array,key)
print("Position of the element linear search",pos_linear)
print("Position of the element binary search",pos_binary)
linear_time = []
binary_time = []
inputs = [1,250,420,630,997]
for i in inputs:
    start_time = time.time()
    linear_search(array,i)
    end_time = time.time()
    elapsed_time = end_time - start_time
    linear_time.append(elapsed_time)
for i in inputs:
    start_time = time.time()
    binary_search(array,i)
    end_time = time.time()
    elapsed_time = end_time - start_time
    binary_time.append(elapsed_time)
plt.plot(inputs, linear_time, label='Linear Search')
plt.plot(inputs, binary_time, label='Binary Search')
plt.xlabel('Search Key')
plt.ylabel('Time (seconds)')
plt.title('Time Taken for Linear and Binary Searches')
plt.legend()
plt.show()
```

```python
#Q3
def string_to_integer(s):
    if not s:
        return 0
    else:
        digit = int(s[0])
        return digit * (10 ** (len(s) - 1)) + string_to_integer(s[1:])

string1 = "13531"
string_to_integer(string1)
```

```python
#Q4
def reverse_string(s):
    if len(s) == 1:
        return s
    else:
        return s[-1] + reverse_string(s[:-1])

original_string = "pots&pans"
reversed_string = reverse_string(original_string)
print("Original String:", original_string)
print("Reversed String:", reversed_string)
```
